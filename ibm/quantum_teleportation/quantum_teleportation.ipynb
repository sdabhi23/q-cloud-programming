```python
%matplotlib inline

from qiskit import QuantumCircuit, execute, BasicAer
from qiskit.visualization import *
import random
```

```python
simulator = BasicAer.get_backend('statevector_simulator')

circuit = QuantumCircuit(3, 3)

ranX = random.random()
ranY = random.random()
circuit.rx(ranX, 0)
circuit.ry(ranY, 0)

circuit.draw('mpl')
```

```python
job = execute(circuit, simulator, shots=4086)
result = job.result()
```

```python
print(result.get_statevector())
plot_bloch_multivector(result.get_statevector())
```

```python
plot_state_city(result.get_statevector())
```

```python
counts = result.get_counts()
plot_histogram(counts)
```

```python
circuit.h(1)
circuit.cnot(1, 2)

circuit.cnot(0, 1)
circuit.h(0)

circuit.measure([0, 1], [0, 1])

circuit.cnot(1, 2)
circuit.cz(0, 2)

circuit.draw('mpl')
```

```python
job = execute(circuit, simulator, shots=4086)
result = job.result()
```

```python
print(result.get_statevector())
plot_bloch_multivector(result.get_statevector())
```

```python
plot_state_city(result.get_statevector())
```

```python
counts = result.get_counts()
plot_histogram(counts)
```
